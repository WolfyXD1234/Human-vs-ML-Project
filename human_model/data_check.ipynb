```python
import pandas as pd
import matplotlib.pyplot as plt

Wine = pd.read_csv("/workspaces/Human-vs-ML-Project/data/Wine.csv")

for x in Wine.columns.tolist():
    fig, ax = plt.subplots()
    ax.scatter(Wine[x],Wine["color"],c=Wine["color"], label = Wine["color"])
    plt.xlabel(x)
    ax.patch.set_facecolor('gray')
    fig.patch.set_facecolor('gray')
    plt.show()
```
